```python
X = df.drop(columns="DODRace",axis=1)
y = df["DODRace"]
```

```python
from sklearn.model_selection import train_test_split 
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=101,stratify=y,test_size=0.25)
```

```python
from sklearn.compose import make_column_transformer
```

```python
df.select_dtypes("object").columns
```

```python
ord_enc_list = ["Gender","SubjectsBirthLocation","WritingPreference"]
```

```python
ord = OrdinalEncoder(handle_unknown="use_encoded_value",unknown_value=-1)
```

```python
column_trans = make_column_transformer((ord,ord_enc_list),remainder="passthrough")
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
operations = [("OrdEncoder",column_trans),("classifier",DecisionTreeClassifier())]
```

```python
model = Pipeline(steps=operations)
```

```python
model.fit(X_train,y_train)
```

```python
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
```

```python
print(confusion_matrix(y_test, y_test_pred))
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
print(confusion_matrix(y_train, y_train_pred))
```

```python
print(classification_report(y_train, y_train_pred))
```
